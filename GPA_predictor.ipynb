```python
#defining function for Linear regression
def Linear_model(x,y):
    x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.2,random_state=0)
    model=LinearRegression()
    model.fit(x_train,y_train)
    return model

#defining function for SVM
def SVM_model(x,y):
    x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.2,random_state=0)
    model=svm.SVR(C=2, kernel='linear')
    model.fit(x_train,y_train)
    return model
```

```python
#for model 01
dt_x1=data.iloc[:,1:12]
dt_y1=data.iloc[:,-1]
algo_11=Linear_model(dt_x1,dt_y1)
algo_12=SVM_model(dt_x1,dt_y1)
```
